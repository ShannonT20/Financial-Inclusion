```python
#Label encode arbitrarly to sll categorical
from feature_engine.encoding import OrdinalEncoder

oe = OrdinalEncoder(encoding_method='arbitrary')

train[train.select_dtypes(include='object').columns] = oe.fit_transform(train.select_dtypes(include='object'))
test[test.select_dtypes(include='object').columns] = oe.fit_transform(test.select_dtypes(include='object'))
```

```python
#create year_of_birth
train = train.assign(
    year_of_birth = train.year - train.age_of_respondent
)
test = test.assign(
    year_of_birth = test.year - test.age_of_respondent
)
```

```python
train.shape
```

```python
#Convert to datetime features
train[['year','year_of_birth']] = train[['year','year_of_birth']].apply(lambda x:pd.to_datetime(x,format='%Y-%m-%d',errors='coerce'))
test[['year','year_of_birth']] = test[['year','year_of_birth']].apply(lambda x:pd.to_datetime(x,format='%Y-%m-%d',errors='coerce'))

train.dtypes
```

```python
#Create datetime features
from feature_engine.datetime import DatetimeFeatures
dtf = DatetimeFeatures(features_to_extract = ["year", "month", "day_of_month"])
train = dtf.fit_transform(train)
test = dtf.fit_transform(test)
```

```python
#paramemeter for the model
from sklearn.metrics import accuracy_score
import lightgbm as lgb
parameters = {
            'nthread':4,
            'n_estimators':50,
            'learning_rate':0.066,
            'num_leaves':30,
            'colsample_bytree':0.9497036,
            'subsample':0.8715623,
            'max_depth':0,
            'reg_alpha':0,
            'reg_lambda':0.0735294,
            'min_split_gain':0.0222415,
            'min_child_weight':39.3259775,
            'silent':-1,
            'verbose':-1,
            'bagging_seed':5,
            'feature_fraction': 0.957,
            'bagging_fraction': 0.956,
            'boosting_type': 'gbdt',
            'objective': 'binary',
            'metric': 'auc'}


train_data_ = lgb.Dataset(X_train, label=y_train)
test_data_ = lgb.Dataset(X_test, label=y_test)


clf =lgb.train(params=parameters, 
                train_set=train_data_,
                valid_sets=[test_data_])

# Make predictions on the test set
y_pred = clf.predict(X_test, num_iteration=clf.best_iteration)
y_pred_binary = (y_pred > 0.5).astype(int)


# Calculate the error rate
error_rate = 1 - accuracy_score(y_test, y_pred_binary)
print(f'Error rate of CatBoost classifier: {error_rate:.4f}')
```
